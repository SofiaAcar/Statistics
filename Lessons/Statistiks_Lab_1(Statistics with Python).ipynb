# Population and Sample

```python
import numpy as np
```

Aynı random sayıları üretmek için seed methodu kullanılır ve parantez içine aynı sayı  yazılır

```python
np.random.seed(101)   
population = np.random.randint(0, 100, size=100000)   #1-100 arası rastgele sayılar oluşturduk
```

```python
population
```

```python
len(population)     
```

```python
np.random.seed(51)     
sample = np.random.choice(population, 100)
```

Yukarıda hocayla aynı rastgele sayılar oluştu. Çünkü seed methodu kullanıp aynı sayıları kullandık.

```python
sample
```

```python
population.mean()  #Population un ortalamasını aldı (Seçim sonucu örnek veerilebilir)
```

```python
sample.mean()  #Sample ın ortalaması. Populasyona yakın bir ort. (Seçim tahmini örnek verilebilir)
```

100 tane sample ın ortalamasını bulmak istiyoruz diyelim:

```python
sample_means =  []
for i in range(100):
    sample = np.random.choice(population, 100)
    sample_means.append(sample.mean())
sample_means
```

Her seferinde rastegele seçilen 100 kişi aynı olmayacağı için her seferinde farklı  sonuç çıkar. Beklenen ort değer populasyonun ortalamasıdır. Yani burda 50 dir.

Sample ın ort. sı da population ort. sına yakın çıktı. Sample bizim tahminleyicimizdir. Sample size' ı sayısal olarak  yükseltsek daha sağlıklı sonuç elde ederiz çünkü veri çok.

```python
np.mean(sample_means)
```
