# Population and Sample

```python
import numpy as np
```

AynÄ± random sayÄ±larÄ± Ã¼retmek iÃ§in seed methodu kullanÄ±lÄ±r ve parantez iÃ§ine aynÄ± sayÄ±  yazÄ±lÄ±r

```python
np.random.seed(101)   
population = np.random.randint(0, 100, size=100000)   #1-100 arasÄ± rastgele sayÄ±lar oluÅŸturduk
```

```python
population
```

```python
len(population)     
```

```python
np.random.seed(51)     
sample = np.random.choice(population, 100)
```

YukarÄ±da hocayla aynÄ± rastgele sayÄ±lar oluÅŸtu. Ã‡Ã¼nkÃ¼ seed methodu kullanÄ±p aynÄ± sayÄ±larÄ± kullandÄ±k.

```python
sample
```

```python
population.mean()  #Population un ortalamasÄ±nÄ± aldÄ± (SeÃ§im sonucu Ã¶rnek veerilebilir)
```

```python
sample.mean()  #Sample Ä±n ortalamasÄ±. Populasyona yakÄ±n bir ort. (SeÃ§im tahmini Ã¶rnek verilebilir)
```

100 tane sample Ä±n ortalamasÄ±nÄ± bulmak istiyoruz diyelim:

```python
sample_means =  []
for i in range(100):
    sample = np.random.choice(population, 100)
    sample_means.append(sample.mean())
sample_means
```

Her seferinde rastegele seÃ§ilen 100 kiÅŸi aynÄ± olmayacaÄŸÄ± iÃ§in her seferinde farklÄ±  sonuÃ§ Ã§Ä±kar. Beklenen ort deÄŸer populasyonun ortalamasÄ±dÄ±r. Yani burda 50 dir.

Sample Ä±n ort. sÄ± da population ort. sÄ±na yakÄ±n Ã§Ä±ktÄ±. Sample bizim tahminleyicimizdir. Sample size' Ä± sayÄ±sal olarak  yÃ¼kseltsek daha saÄŸlÄ±klÄ± sonuÃ§ elde ederiz Ã§Ã¼nkÃ¼ veri Ã§ok.

```python
np.mean(sample_means)
```
